```python
import os
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from mido import MidiFile
import unidecode
from midiutil.MidiFile import MIDIFile
```

```python
base_dir = 'LSTMImages1'
midi_folders = os.listdir(base_dir)
```

```python
midi_names = []
for folder in midi_folders:
    midi_loc = os.path.join(base_dir, folder)
    for s in os.listdir(midi_loc):
        midi_names.append(folder +'\\'+ s)
```

```python
pixel_sec = 0.05
tempo = 500000.0
division = 480.0
img_heigth = 88
min_note = 21
```

```python
def getLength(a_img, x, y):
    i = x
    j = y + 1
    while(a_img[i][j] != 0 and a_img[i][j] != 125):
        j += 1
    return (j - y)*pixel_sec
```

```python
def findNote(a_img, i, j):
    if(i >= a_img.shape[0]):
        i = 0
        j += 1
    for x in range(j,a_img.shape[1]):
        for y in range(i, a_img.shape[0]):
            if(a_img[y][x] == 125):
                length = getLength(a_img, y, x)
                time = int(np.ceil(1000000*pixel_sec*x*division/tempo))
                dur = int(np.ceil(1000000*length*division/tempo))
                print(y, x)
                return (min_note + 87 - y), dur, time, y, x
    return -1, -1, -1, -1, -1
```

```python
for file_name in midi_names:
    img = cv2.imread('c:/NHF/LSTMimages1/' + str(file_name), cv2.IMREAD_UNCHANGED)
    print(img.shape)
    # create your MIDI object
    mf = MIDIFile(numTracks=1, ticks_per_quarternote=480, eventtime_is_ticks=True)
    track = 0  

    time = 0    # start at the beginning
    mf.addTrackName(track, time, "Sample Track")
    mf.addTempo(track, time, 120)
    
    track = 0
    channel = 0
    program = 0
    volume = 100
    mf.addProgramChange(track, channel, time, program)
    note, dur, time, n, w = findNote(img, 0, 0)
    while(note != -1):   
        mf.addNote(track, channel, note, time, dur, volume)
        note, dur, time, n, w = findNote(img, n + 1, w)

# write it to disk
with open("output.mid", 'wb') as outf:
    mf.writeFile(outf)
```
